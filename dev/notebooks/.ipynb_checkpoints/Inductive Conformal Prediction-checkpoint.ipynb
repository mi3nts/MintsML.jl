Let's do a simple example. Let the true values given by $y=sin(x)$ and a model given by the 3rd order taylor expansion $f(x) = x - x^3/3!$ 

```julia
x = 
y = sin.(x) 

f(x) = x - x^3/6 

f̂ = f.(x)

plot(x, y, seriestype=:scatter, color=:red, label="true")
plot!(x, f̂, color=:lightblue, label="model")
```
